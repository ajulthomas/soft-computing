# Regression - California Housing Dataset


The data has the following features: longitude, Latitude, housing_median_age, total_rooms, total_bedrooms, population, households, median_income, ocean_proximity. 


The target is: median_house_value

```python
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
X, y = california.data, california.target
feature_names = california.feature_names
```

```python
print(feature_names)
```

```python
X.shape
```

```python
y.shape
```

```python
import matplotlib.pyplot as plt

for i in range(len(feature_names)):
  fig,ax = plt.subplots()
  ax.scatter(X[:,i], y)
  ax.set_xlabel(feature_names[i])
  ax.set_ylabel(california.target_names[0])
  fig.show()
```

```python
from sklearn.linear_model import LinearRegression

for j in range(len(feature_names)):
  simple_linear_reg_model = LinearRegression().fit(X[:, j].reshape(-1, 1), y)
  y_pred = simple_linear_reg_model.predict(X[:, j].reshape(-1, 1))
  MSE = ((y - y_pred)** 2).mean()
  print("MSE of simple linear regression using feature " , feature_names[j] , MSE)
```

```python
multiple_linear_reg_model = LinearRegression().fit(X, y)
y_pred = multiple_linear_reg_model.predict(X)
MSE = ((y - y_pred)** 2).mean()
print("MSE of multiple linear regression using all features:" ,  MSE)
```

# Classification

The dataset contains the following columns: Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, and Species. 

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
X, y = iris.data, iris.target
feature_names_iris = iris.feature_names
```

```python
print(feature_names_iris)
```

```python
for i in range(len(feature_names_iris)):
  fig,ax = plt.subplots()
  ax.scatter(X[:,i], y, c=iris.target, cmap='viridis')
  ax.set_xlabel(feature_names_iris[i])
  ax.set_ylabel(iris.target_names[0])
  fig.show()
```

```python
from sklearn.linear_model import LogisticRegression

for j in range(len(feature_names_iris)):
    simple_logistic_reg_model = LogisticRegression().fit(X[:, j].reshape(-1, 1), y)
    y_pred = simple_logistic_reg_model.predict(X[:, j].reshape(-1, 1))
    MSE = ((y - y_pred)** 2).mean()
    accuracy = (y == y_pred).mean()
    print("MSE of simple logistic regression using feature " , feature_names_iris[j] , MSE)
    print("Accuracy of simple logistic regression using feature " , feature_names_iris[j] , accuracy)
```

```python
iris_logistic_reg_model = LogisticRegression(solver='lbfgs', max_iter=1000).fit(X, y)
y_pred = iris_logistic_reg_model.predict(X)
accuracy = (y == y_pred).mean()
MSE = ((y - y_pred)** 2).mean()
print("MSE of multiple logistic regression using all features:" ,  MSE)
print("Accuracy of multiple logistic regression using all features:" ,  accuracy)
```
