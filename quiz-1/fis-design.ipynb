# Design of Fuzzy Interface System for Robot Experiment

**Selected task**
Design a fuzzy interface system to predict the velocity of the robot at any given time, based on the inputs given below:

## Define Inputs and Outputs

**inputs**

- distance_x_candle_1 (x distance of the robot to candle 1)
- distance_y_candle_1 (x distance of the robot to candle 1)

- distance_x_candle_2 (x distance of the robot to candle 1)
- distance_y_candle_2 (x distance of the robot to candle 1)

- distance_x_candle_3 (x distance of the robot to candle 1)
- distance_y_candle_3 (x distance of the robot to candle 1)

- distance_x_candle_4 (x distance of the robot to candle 1)
- distance_y_candle_4 (x distance of the robot to candle 1)

- collision

- orientation_candle_1 (archtan of distance x and distance y)
- orientation_candle_2 (archtan of distance x and distance y)
- orientation_candle_3
- orientation_candle_4

**output**

- velocity X
- velocity Y

**Assumptions**

- Assumes that the robot has a light detector and ultrasonic sensor to detect proximity to other objects or walls
- Assumes that the accelartaion of the robot is constant
- assumes that the orientation is the orientation of the light detector
- Assumes that the robot is guided by the light detector and orientation represents the orientation of the light sensor
- Assumes that the robot accelarates at constant rate towards the direction of the light
- incase of collision or detecting a wall the velocity reduces
- if distance is high velocity is high
- Assumes that the size of the box is 100 metre square
- assumes that the maximum velocity of the robots is 1 m/s
- assumes that the orientation is measured in degrees
- assumes that

## Design Knowledge Base

- **Rule 1**:
  if
  (distance_x_candle_1 is high AND
  distance_y_candle_1 is high AND
  orientation_candle_1 is high AND
  collision is low)
  OR
  (distance_x_candle_2 is high AND
  distance_y_candle_2 is high AND
  orientation_candle_2 is high AND
  collision is low)
  OR
  (distance_x_candle_3 is high AND
  distance_y_candle_3 is high AND
  orientation_candle_3 is high AND
  collision is low)
  OR
  (distance_x_candle_4 is high AND
  distance_y_candle_4 is high AND
  orientation_candle_4 is high AND
  collision is low)
  then
  velocity_x is high
  velocity_y is high

- **Rule 2**

  if
  (distance_x_candle_1 is low AND
  distance_y_candle_1 is low AND
  orientation_candle_1 is high AND
  collision is low)
  OR
  (distance_x_candle_2 is low AND
  distance_y_candle_2 is low AND
  orientation_candle_2 is high AND
  collision is low)
  OR
  (distance_x_candle_3 is low AND
  distance_y_candle_3 is low AND
  orientation_candle_3 is high AND
  collision is low)
  OR
  (distance_x_candle_4 is low AND
  distance_y_candle_4 is low AND
  orientation_candle_4 is high AND
  collision is low)
  then
  velocity_x is medium
  velocity_y is medium


# Design Inference Engine

If the collision is high, the velocity will always be low
if orientation is high, the velocity will be either high or low

AND is computed as minimum
OR is computed as maximum
